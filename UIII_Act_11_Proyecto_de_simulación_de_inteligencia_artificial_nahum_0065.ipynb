```python
# =========================================================
# SISTEMA IA - CONTROL DE ADMINISTRADORES
# CBTIS 128
# Carrera: Inteligencia Artificial
# Materia: Elabora Proyectos de Programación Lógica
# Alumno: Nahum Flores
# NL = 21
# =========================================================

print("===================================================")
print("     IA EMPRESARIAL - CONTROL DE ADMINISTRADORES")
print("===================================================")

# Lista de diccionarios vacía
lista_administradores = []

# Diccionario vacío
administradores = {}

# =====================================================
# FUNCIÓN PARA REGISTRAR ADMINISTRADORES
# =====================================================

def registrar_administrador():

    global administradores

    print("\n===== REGISTRO DE ADMINISTRADOR =====")

    nombre = input("Nombre del administrador: ")
    empresa = input("Empresa: ")
    departamento = input("Departamento: ")
    presupuesto = input("Presupuesto asignado: ")
    personal = input("Personal a cargo: ")
    metas = input("Metas trimestrales: ")
    software = input("Software ERP utilizado: ")

    # Diccionario
    administradores = {
        "nombre": nombre,
        "empresa": empresa,
        "departamento": departamento,
        "presupuesto": presupuesto,
        "personal": personal,
        "metas": metas,
        "software": software
    }

    # Guardar en la lista
    lista_administradores.append(administradores)

    print("\n IA: Administrador registrado correctamente")

    # Inteligencia artificial simple
    personal_int = 0
    if personal.strip():
        try:
            personal_int = int(personal.strip())
        except ValueError:
            print(" IA AVISO: Valor de personal no válido, asumiendo 0 para evaluación.")
            personal_int = 0

    if personal_int >= 50:

        print(" IA EMPRESARIAL: Administrador de alto nivel")

    elif personal_int >= 20:

        print(" IA: Administrador con nivel intermedio")

    else:

        print(" IA AVISO: Administrador con equipo pequeño")


# =====================================================
# FUNCIÓN PARA MOSTRAR ADMINISTRADORES
# =====================================================

def mostrar_administradores():

    print("\n================ LISTA DE ADMINISTRADORES ================")

    if len(lista_administradores) == 0:

        print("No existen administradores registrados")

    else:

        print("-" * 130)

        print("{:<15} {:<15} {:<20} {:<20} {:<18} {:<22} {:<15}".format(
            "NOMBRE",
            "EMPRESA",
            "DEPARTAMENTO",
            "PRESUPUESTO",
            "PERSONAL",
            "METAS",
            "SOFTWARE ERP"
        ))

        print("-" * 130)

        for datos in lista_administradores:

            print("{:<15} {:<15} {:<20} {:<20} {:<18} {:<22} {:<15}".format(
                datos["nombre"],
                datos["empresa"],
                datos["departamento"],
                datos["presupuesto"],
                datos["personal"],
                datos["metas"],
                datos["software"]
            ))

        print("-" * 130)


# =====================================================
# FUNCIÓN PARA BUSCAR ADMINISTRADOR
# =====================================================

def buscar_administrador():

    buscar = input("\nEscribe el nombre del administrador: ")

    encontrado = False

    for datos in lista_administradores:

        if datos["nombre"].lower() == buscar.lower():

            print("\n========= ADMINISTRADOR ENCONTRADO =========")
            print("Nombre:", datos["nombre"])
            print("Empresa:", datos["empresa"])
            print("Departamento:", datos["departamento"])
            print("Presupuesto asignado:", datos["presupuesto"])
            print("Personal a cargo:", datos["personal"])
            print("Metas trimestrales:", datos["metas"])
            print("Software ERP:", datos["software"])

            encontrado = True

    if encontrado == False:

        print(" IA: Administrador no encontrado")


# =====================================================
# FUNCIÓN IA DE EVALUACIÓN
# =====================================================

def evaluar_administradores():

    print("\n=========== EVALUACIÓN IA EMPRESARIAL ===========")

    if len(lista_administradores) == 0:

        print("No existen administradores registrados")

    else:

        for datos in lista_administradores:

            print("\nAdministrador:", datos["nombre"])
            presupuesto_int = 0
            if datos["presupuesto"].strip():
                try:
                    presupuesto_int = int(datos["presupuesto"].strip())
                except ValueError:
                    print(" IA AVISO: Valor de presupuesto no válido, asumiendo 0 para evaluación.")
                    presupuesto_int = 0

            if presupuesto_int >= 1000000:

                print(" IA: Maneja presupuesto empresarial alto")

            elif presupuesto_int >= 500000:

                print(" IA: Presupuesto empresarial medio")

            else:

                print(" IA AVISO: Presupuesto empresarial bajo")


# =====================================================
# MENÚ PRINCIPAL
# =====================================================

while True:

    print("\n================ MENÚ PRINCIPAL ================")
    print("1. Registrar administrador")
    print("2. Mostrar administradores")
    print("3. Buscar administrador")
    print("4. Evaluación IA empresarial")
    print("5. Salir")

    opcion = input("Selecciona una opción: ")

    # =================================================
    # OPCIÓN 1
    # =================================================

    if opcion == "1":

        registrar_administrador()

    # =================================================
    # OPCIÓN 2
    # =================================================

    elif opcion == "2":

        mostrar_administradores()

    # =================================================
    # OPCIÓN 3
    # =================================================

    elif opcion == "3":

        buscar_administrador()

    # =================================================
    # OPCIÓN 4
    # =================================================

    elif opcion == "4":

        evaluar_administradores()

    # =================================================
    # OPCIÓN 5
    # =================================================

    elif opcion == "5":

        print("\n===================================")
        print(" Cerrando sistema empresarial IA")
        print(" Gracias por utilizar el programa")
        print("===================================")

        break

    # =================================================
    # OPCIÓN INVÁLIDA
    # =================================================

    else:

        print("\n Opción inválida")
        print(" Intenta nuevamente")
```
